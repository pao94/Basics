```python
# Define a function that returns True for leap years, else False

def is_leap(year):
    return (year % 4 == 0 and not year % 100 == 0) or year % 400 == 0

is_leap(2000)

# "return condition" evaluastes a boolean. If-statements in this case
```

```python
# Given a number n, print 123...n as a string without spaces. E.g., n = 5, print "12345"

n = 5

print(*range(1,int(n)+1), sep ="") 

# * unpacks an iterator e.g., print(*range(1,5)) outputs: 1 2 3 4
# if we were to do print(range(1,5)) the output would be: range(1,5)
# another example: print([1,2,3]) outputs [1,2,3] while print(*[1,2,3]) outputs 1 2 3
```

```python
# Given n scores and an array of n integers separated by a space, print the runner up score

if __name__ == '__main__':
    n = int(input())
    arr = map(int, input().split())
    
    print(sorted(set(arr))[-2])

# Sets cannot include duplicates, which takes care of multiple maximums. Sorted() returns a ascending list of values, taking second last prints the runner up.
```

```python
# Given first line with integer n, the number of student's record, the next lines n containing names and grades of each student with each value seperated by a space, and the final line containing query_name with name of student to query, print the average grade of said student rounded to two decimal places.

# From each line containing names and grades, we create a dictionary entry with name as key and grades as value pairs. E.g., name:[marks]

if __name__ == '__main__':
    n = int(input())
    student_marks = {}
    for _ in range(n):
        name, *line = input().split()
        scores = list(map(float, line))
        student_marks[name] = scores
    query_name = input()
    
    marks = student_marks.get(query_name)
    
    print(f"{sum(marks)/len(marks):.2f}")

# again we use * in "name, *line" because it unpacks the iterator returned by input().split() which is ["Krishna", "67", "68", "69"]
# input() returns a string, while split() returns a list
# name, line = input().split() wouldnt work because it would only expect two values (one for name, one for line)
```

```python
# Given names and grades of each student, store them in a nested list and print the name(s) of any students having the second lowest grade on a new line for each one

if __name__ == '__main__':
    
    n = int(input())
    python_students = [[input(), float(input())] for _ in range(n)]
        
    second = sorted(set([marks for names, marks in python_students]))[1]
    print(*sorted([names for names, marks in python_students if marks == second]), sep = "\n")
```

```python
# Given integer n, denoting total number of country stamps and next lines N containing countries where stamp is from, print distinct number of country stamps

if __name__ == '__main__':
    n = int(input())
    print(len({input() for _ in range(n)}))
```

```python
def bubbleSort(arr):
    n = len(arr)

    for i in range(n-1):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                
    return arr
```
